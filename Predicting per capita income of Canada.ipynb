```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("C:\\Users\\ARNAB DAS\\Desktop\\canada's bla bla.csv",skiprows=1)
df
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Per capita income")
plt.scatter(df.year,df.pci, color="red")
```

```python
# Now declare an attribute from sklearn and that is LinearRegression()
reg = linear_model.LinearRegression()
#We want to predict per capita income wrt year...
reg.fit(df[['year']],df.pci)
#this model is complete for predicting
```

```python
#Now i want to predict the per capita income in year 2020
reg.predict([[2020]])
```

```python
# For getting the regression line
%matplotlib inline
plt.xlabel("YEARS")
plt.ylabel("PER CAPITA INCOME")
plt.scatter(df.year,df.pci, color="red")
# For plotting the regression line
plt.plot(df.year,reg.predict(df[['year']]),color='blue') # We should only predict the pci values wrt year
```
